```python
from pycaret.regression import *
from sklearn.datasets import fetch_california_housing
import pandas as pd

housing = fetch_california_housing(as_frame=True)
df = housing.frame

reg_setup = setup(
    data=df,
    target='MedHouseVal', 
    session_id=42,
    normalize=True,
    verbose=False,
    log_experiment=False 
)

best_model = compare_models()
print("Best model:", best_model)
final_model = finalize_model(best_model)

new_data = pd.DataFrame({
    'MedInc': [8.3252, 5.6431],
    'HouseAge': [41, 21],
    'AveRooms': [6.9841, 6.2381],
    'AveBedrms': [1.0238, 0.9719],
    'Population': [322, 240],
    'AveOccup': [2.5556, 2.1098],
    'Latitude': [37.88, 37.86],
    'Longitude': [-122.23, -122.22]
})

predictions = predict_model(final_model, data=new_data)
predictions['Predicted_Value_USD'] = predictions['prediction_label'] * 100000
print(predictions[['Predicted_Value_USD']])
```
